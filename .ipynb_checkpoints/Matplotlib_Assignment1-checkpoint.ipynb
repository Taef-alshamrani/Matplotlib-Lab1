```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q2: Create pie chart with equal explode values.

```python
x=[10, 20, 30, 40]
plt.pie(x, autopct='%1.1f%%', explode = [0.1, 0.1, 0.1, 0.1]);
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.scatter(x,y, marker='o', s = 50, alpha = 0.4);
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.scatter(x,y, marker='o', s = 50, alpha = 0.4)
plt.title('Q4', color='r', fontsize = 25, loc='left')
plt.xlabel('linspace', fontsize = 16)
plt.ylabel('sin', fontsize = 16);
```

## Q5: Create three line charts with Legends at left and colors.

```python
plt.plot(x, np.sin(x - 0), color='blue')        
plt.plot(x, np.sin(x - 1), color='g')           
plt.plot(x, np.sin(x - 2), color='0.75') 
plt.legend(['sin(x - 0)','sin(x - 1)', 'sin(x - 2)'], shadow = True, loc = 'upper left');
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
plt.figure(figsize=[20,10])
plt.suptitle('Q7')


plt.subplot(1,3,1)
x=[10, 20, 30, 40]
plt.title('From Q2', color='r', fontsize = 25, loc='left')
plt.pie(x, autopct='%1.1f%%', explode = [0.1, 0.1, 0.1, 0.1]);

plt.subplot(1,3,2)
x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.scatter(x,y, marker='o', s = 50, alpha = 0.4)
plt.title('From Q4', color='r', fontsize = 25, loc='left')
plt.xlabel('linspace', fontsize = 16)
plt.ylabel('sin', fontsize = 16);

plt.subplot(1,3,3)
plt.plot(x, np.sin(x - 0), color='blue')        
plt.plot(x, np.sin(x - 1), color='g')           
plt.plot(x, np.sin(x - 2), color='0.75') 
plt.title('From Q5', color='r', fontsize = 25, loc='left')
plt.legend(['sin(x - 0)','sin(x - 1)', 'sin(x - 2)'], shadow = True, loc = 'upper left');
```
